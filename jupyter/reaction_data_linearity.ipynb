```python
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
from reaction_model import *

csv_fetcher = CSVFileDataFetcher('training/markov_line_utf8.csv')

rows = csv_fetcher.get_data()

rows_filtered = []
for row in rows:
    if row[0] != '':
        rows_filtered.append(row)

y_reaction = []
y_not_reaction = []

for row_idx, row in enumerate(rows_filtered):

    if int(row[0]) == 0:
        y_not_reaction.append(AOLReactionFeatureAnalyzer.analyze(row[1]))
    else:
        if len(row[1]) > 10:
            print(row[1])
        y_reaction.append(AOLReactionFeatureAnalyzer.analyze(row[1]))

# Convert to NP arrays
y_reaction = np.array(y_reaction)
y_not_reaction = np.array(y_not_reaction)

features = AOLReactionFeatureAnalyzer.features()

for feature_idx, feature in enumerate(features):
    _x_reaction = np.arange(0, len(y_reaction), 1)
    _y_reaction = np.array(y_reaction[:,feature_idx])
    marker = u'+'
    plt.scatter(_x_reaction,_y_reaction,marker=marker, label='reaction')

    _x_not_reaction = np.arange(len(y_reaction), len(y_not_reaction) + len(y_reaction), 1)
    _y_not_reaction = np.array(y_not_reaction[:,feature_idx])
    marker = u'.'
    plt.scatter(_x_not_reaction,_y_not_reaction,marker=marker, label='not reaction')

    plt.legend(loc=2)
    plt.ylabel(feature)
    
    plt.show()
    plt.clf()
```
